Understanding patterns and relationships in house prices.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
sns.histplot(df["price_lkr"], bins=30, kde=True)
plt.title("House Price Distribution (LKR)")
plt.show()
```

```python
sns.boxplot(x="bedrooms", y="price_lkr", data=df)
plt.title("Bedrooms vs Price")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x="perch", y="price_lkr", data=df)
plt.title("Perch vs Price")
plt.show()

sns.boxplot(x="floors", y="price_lkr", data=df)
plt.title("Floors vs Price")
plt.show()

# Pair plot for numerical features
numerical_cols = ['perch', 'bedrooms', 'bathrooms', 'kitchen_area_sqft', 'parking_spots', 'floors', 'house_age', 'price_lkr']
sns.pairplot(df[numerical_cols].sample(1000))  # Sample for speed
plt.show()

# Categorical analysis
plt.figure(figsize=(12,6))
sns.barplot(x="district", y="price_lkr", data=df, estimator=np.mean)
plt.xticks(rotation=90)
plt.title("Average Price by District")
plt.show()
```
