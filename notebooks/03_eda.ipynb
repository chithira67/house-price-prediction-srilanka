Understanding patterns and relationships in house prices.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
print("Available columns:", df.columns.tolist()[:20])
print("Shape:", df.shape)
```

```python
sns.histplot(df["price_lkr"], bins=30, kde=True)
plt.title("House Price Distribution (LKR)")
plt.savefig("../plots/price_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x="bedrooms", y="price_lkr", data=df)
plt.title("Bedrooms vs Price")
plt.savefig("../plots/bedrooms_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.savefig("../plots/correlation_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.scatterplot(x="perch", y="price_lkr", data=df, alpha=0.6)
plt.title("Perch vs Price")
plt.savefig("../plots/perch_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x="floors", y="price_lkr", data=df)
plt.title("Floors vs Price")
plt.savefig("../plots/floors_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
numerical_cols = ['perch', 'bedrooms', 'bathrooms', 'kitchen_area_sqft', 'parking_spots', 'floors', 'house_age', 'price_lkr']
pair_plot = sns.pairplot(df[numerical_cols].sample(1000), diag_kind='kde', plot_kws={'alpha':0.6})
pair_plot.fig.suptitle("Pair Plot of Numerical Features", y=1.02)
plt.savefig("../plots/pair_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python

district_cols = [col for col in df.columns if col.startswith('district_')]
df_plot = df.copy()
df_plot['district'] = df[district_cols].idxmax(axis=1).str.replace('district_', '')

plt.figure(figsize=(12,6))
sns.barplot(x="district", y="price_lkr", data=df_plot, estimator=np.mean)
plt.xticks(rotation=90)
plt.title("Average Price by District")
plt.savefig("../plots/average_price_by_district.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.histplot(df["house_age"], bins=20, kde=True)
plt.title("House Age Distribution")
plt.savefig("../plots/house_age_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x="bathrooms", y="price_lkr", data=df)
plt.title("Bathrooms vs Price")
plt.savefig("../plots/bathrooms_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x="parking_spots", y="price_lkr", data=df)
plt.title("Parking Spots vs Price")
plt.savefig("../plots/parking_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.scatterplot(x="kitchen_area_sqft", y="price_lkr", data=df, alpha=0.6)
plt.title("Kitchen Area vs Price")
plt.savefig("../plots/kitchen_area_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python

district_cols = [col for col in df.columns if col.startswith('district_')]
df_plot = df.copy()
df_plot['district'] = df[district_cols].idxmax(axis=1).str.replace('district_', '')

plt.figure(figsize=(12,6))
sns.countplot(x="district", data=df_plot)
plt.xticks(rotation=90)
plt.title("Number of Houses by District")
plt.savefig("../plots/houses_by_district.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x="has_garden", y="price_lkr", data=df)
plt.title("Price by Garden Presence")
plt.savefig("../plots/price_by_garden.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(x="has_ac", y="price_lkr", data=df)
plt.title("Price by AC Presence")
plt.savefig("../plots/price_by_ac.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.scatterplot(x="total_rooms", y="price_lkr", data=df, alpha=0.6)
plt.title("Total Rooms vs Price")
plt.savefig("../plots/total_rooms_vs_price.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.histplot(df["luxury_score"], bins=5, kde=False)
plt.title("Luxury Score Distribution")
plt.savefig("../plots/luxury_score_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```
